```python
import re
import pandas as pd
import json
from collections import OrderedDict
from nltk.corpus import stopwords

class pembersihan:
    @staticmethod
    def clear_word(caption):
        # casefolding
        caption = caption.lower()
        
        # remove angka
        caption = re.sub('[0-9]+', '', caption)

        # remove hyperlinks
        caption = re.sub(r'https?:\/\/\S+', '', caption)

        # remove hashtags
        caption = re.sub(r'#[A-Za-z0-9_]+', '', caption)

        # remove symbol
        caption = re.sub(r'[^\x00-\x7f]', '', caption)
        
        # remove mention and link
        caption = re.sub(r"(?:\@|http?\://|https?\://|www)\S+", "", caption)

        # remove coma
        caption = re.sub(r',',' ', caption)

        # remove double huruf
        caption = ''.join(OrderedDict.fromkeys(caption))

        # remove punctuation
        caption = re.sub(r'[^\w\s]', '', caption)

        return caption
    
    @staticmethod
    def remove_sentence(caption): 
        word = caption.split()
        wordCount = len(word)
        if(wordCount<=1):
            caption = ''
        return caption
    
    @staticmethod
    def slang_word_dict(caption):
        df_slang = pd.read_csv('colloquial-indonesian-lexicon.csv')
        slang_dict = dict(zip(df_slang['slang'], df_slang['formal']))
        return [slang_dict.get(token, token) for token in caption.split()]

    @staticmethod
    def slang_word_json(caption):
        with open('src/combined_slang_words.txt', 'r') as f:
            slangword = json.load(f)
        return [slangword.get(token, token) for token in caption.split()]

    @staticmethod
    def stopwords(caption):
        stop_words = set(stopwords.words('indonesian'))
        stop_words.update([
             'min', 'ga', 'gak', 'iya', 
            'ya', 'sih', 'gk', 'kak', 
            'bang', 'pak', 'sir', 'bro', 
            'sob', 'sdr','mbak', 'mba', 
            'kakak', 'kakaknya', 'kakakku', 
            'nya', 'mas'])
        return [word for word in caption.split() if not word in stop_words]
```
