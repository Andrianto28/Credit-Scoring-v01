```python
from transformers import pipeline
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
```

```python
pretrained= "mdhugol/indonesia-bert-sentiment-classification"
model = AutoModelForSequenceClassification.from_pretrained(pretrained)
tokenizer = AutoTokenizer.from_pretrained(pretrained)
```

```python
sentiment_analysis = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
label_index = {'LABEL_0': 'Positif', 'LABEL_1': 'Netral', 'LABEL_2': 'Negatif'}
```

```python
def get_sentiment_indobert(text):
    result = sentiment_analysis(text)
    status = label_index[result[0]['label']]
    return status
```

```python
#bersih bersih
commentar['bersih'] = commentar['comment_text'].apply(lambda x: pb.Pembersihan.clear_word(x))
commentar['bersih'] = commentar['bersih'].apply(lambda x: pb.Pembersihan.remove_sentence(x))
```

```python
commentar['sentiment_indobert'] = commentar['bersih'].apply(get_sentiment_indobert) 
```

```python
# Menghitung jumlah sentiment untuk setiap user
sentiment_count = commentar.groupby(['username', 'sentiment_indobert'])['id_user'].count().reset_index(name='count')

# Mencari sentiment terbanyak berdasarkan user
max_sentiment = sentiment_count.groupby(['username'])['count'].transform(max) == sentiment_count['count']
most_sentiment_comment = sentiment_count[max_sentiment]

# Menampilkan hasil
print(most_sentiment_comment)
```

## Scoring

```python
def scoring_comment(sentiment):
    total = 0
    if sentiment == 'Positif':
        total = 30
    elif sentiment == 'Negatif':
        total = 5
    else:
        total = 15
    
    return total

def scoring_feed(label):
    total = 0
    if label == 'Sarkas/Sindiran':
        total = 5
    elif label == 'Kehidupan Pribadi':
        total = 15
    else :
        total = 30
    return total
        
def scoring_keaktifan(aktif):
    total = 0
    if aktif == 1:
        total = 30
    else:
        total = 10
    return total
```

```python
merged['socring_aktif']= merged['scoring'].apply(lambda x: scoring_keaktifan(x))
merged['socring_feed']= merged['category'].apply(lambda x: scoring_feed(x))
merged['socring_comment']= merged['sentiment_indobert'].apply(lambda x: scoring_comment(x))
```

```python
merged['total_score'] = merged['socring_aktif'] + merged['socring_feed'] + merged['socring_comment']
```

```python
df_total_score = merged.groupby(['id_users', 'username'])['total_score'].sum().reset_index()
```

```python
def kriteria_Kredit(score):
    if score > 75:
        kriteria = 'Sangat Terpercaya'
    elif 60 < score <= 75:
        kriteria = 'Cukup Terpercaya'
    else:
        kriteria = 'Sangat Tidak Terpercaya'
    
    return kriteria
```

```python
df_total_score['Kriteria'] = df_total_score['total_score'].apply(kriteria_Kredit)
```

```python
# mengubah dataframe menjadi JSON
json_data_score = df_total_score.to_json(orient='records')
```
